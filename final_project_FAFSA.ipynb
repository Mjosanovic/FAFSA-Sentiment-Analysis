# Introduction:
Access to higher education poses a significant challenge for low-income and first generation students in the United States.

Despite the existence of financial aid programs like the Free Application for Federal Student Aid (FAFSA), many are discouraged from pursuing higher education. According to the National College Attainment Network (NCAN), 1.7 million students missed out on need-based federal aid in the 2020-21 FAFSA cycle, leaving an estimated *3.7 billion in unclaimed Pell Grants* on the table. Additionally, Forbes reports that 100 million USD in private scholarship funding goes unclaimed each year due to a *lack of applicants*. At the same time, the United States now carries a total of 1,77 trillion USD in student debt. Low-income and first generation students often carry the heaviest weight of this debt.

My research suggests that many of these problems could be solved with higher numbers of completed FAFSA applications. Moreover, a lack of understanding about financial aid leaves many first-generation and low-income students believing they're ineligible, thus not filing the FAFSA.

This lack of understanding and support can discourage individuals from pursuing higher education, further lowering college enrollment rates specifically among low income and first generations groups. The psychological impact of perceiving higher education as unaffordable contributes to a self-fulfilling prophecy, where students from low-income backgrounds may not even consider college as a viable option.

Data supports that there needs to be substantial policy changes to FAFSA, making it more accessible to students who are eligible and deserving of aid.

**Methodology**

In my research, I conducted a sentiment analysis on public opinion regarding the Free Application for Federal Student Aid (FAFSA) over the years 2012, 2013, 2016, and 2024 to get a better understanding of public views on higher education. My aim was to explore whether negative sentiments towards FAFSA remained consistent over the last decade and to create a comparative chart of sentiment over time. To gather the data for this analysis, I employed API-based web scraping tools to extract comments from YouTube videos related to FAFSA. I deliberately chose videos posted directly by FAFSA, rather than those by popular vloggers discussing the same topic. This strategic selection was driven by the hypothesis that comments on FAFSA’s official channel would likely yield more candid feedback. I assumed that commenters, believing their views might be read by FAFSA representatives, would be more inclined to express their genuine sentiments about the financial aid program. This approach aimed to capture a more authentic and direct reflection of public opinion towards FAFSA.

The initial phase of my project involved meticulously cleaning the data. I extracted the comments from YouTube videos and saved them to a TXT file where I manually filtered out comments including non-English words, websites, WhatsApp numbers, and other irrelevant content, such as obvious scams, from the collected text files. This step was crucial to ensure the integrity and relevance of the data for sentiment analysis. After this, I processed the comments further, converting them to lowercase, removing punctuation, lemmatizing, and eliminating stopwords. The cleaned data was then saved in a CSV file for VADER analysis.

One of the primary challenges I faced was in the lemmatization process. I noticed that many words were not fully lemmatized, with instances of words like “running” not being reduced to their base form “run.” To address this issue, I sought assistance from ChatGPT to understand and rectify the shortcomings in my code.

My initial hypothesis was that the comments would predominantly display negative sentiments throughout all four years I was researching. Upon manual review of the comments, this hypothesis appeared to be confirmed, as there were predominantly negative comments. However, this observation starkly contrasted with the results from the VADER sentiment analysis tool. The tool often assigned positive scores to comments, contradicting my findings. This discrepancy primarily arose from VADER's inability to accurately interpret sarcasm: comments with a sarcastic tone, which were inherently negative, were still scored as positive due to the presence of words usually associated with positive sentiments. Thus, while my direct reading of the comments supported my hypothesis of a strong negative sentiment, the sentiment scores provided by VADER did not accurately reflect this, underscoring a significant limitation in relying solely on automated sentiment analysis tools for capturing the true nature of public sentiment. Due to this, I applied BERT scoring to two of the videos. However, the difference was not significant and in some cases even more inaccurate.

The word "joke" at the beginning is probably not strong enough in sentiment scoring to offset the cumulative positive score from the other words precisely because VADER is not able to detect sarcasm.

In each of these examples, the positive score can be attributed to the presence of words that are typically associated with positive sentiments. The tools likely gave more weight to these words, overshadowing the overall negative or sarcastic tone of the comments. This highlights a key limitation in automated sentiment analysis: the challenge of accurately interpreting context, tone, and sarcasm.

*Challenges and limitation*:
1. One of the primary challenges in this research was the significant reduction in data, post-cleaning. Some comment sections were reduced from 70 comments to 30. This raised concerns about the representativeness and comprehensiveness of the analyzed comments. As many comments were filtered out due to language and relevance issues, the remaining dataset might be too small to fully capture the diverse range of sentiments about FAFSA. This limitation highlights the need for a balance between data quality and quantity in sentiment analysis projects.

2. The limitations of the VADER sentiment analysis tool, particularly in its interpretation of sarcasm and nuanced language, presented a significant challenge in my study. Many comments laced with sarcasm were inaccurately scored as positive, highlighting a gap in the tool's ability to decipher complex human expressions. This shortcoming emphasizes the need for more advanced sentiment analysis tools with enhanced capabilities for contextual understanding. To address this, I explored the use of BERT for sentiment analysis in two of the videos, comparing the results with those from VADER through comparative histograms. However, this approach had its own limitations, as I faced uncertainties regarding the accuracy of my implementation of BERT. This exploration underscored the complexities involved in selecting and effectively utilizing advanced sentiment analysis technologies when you are learning Python.

3. Methodological choices, such as selecting specific YouTube videos and manually cleaning data, introduced potential biases. The focus on videos directly from FAFSA's channel might have limited the diversity of perspectives, as viewers of these videos could have pre-existing opinions about FAFSA. This decision also significantly minimized my dataset, creating barriers and reducing the reliability of my findings.

4. Another key limitation in this research was the initial ambition to create a chart representing public sentiment over time, with an aim to investigate trends that might have contributed to fluctuations in the sentiment, particularly the prevalence of negative comments. However, due to the inaccuracies in sentiment scoring, especially the misinterpretation of sarcasm and context by the VADER and BERT tools, the reliability of the data was compromised. Consequently, any chart developed under these conditions would likely have misrepresented the true public sentiment. This realization highlighted the importance of having accurate and reliable sentiment analysis tools before attempting to draw real conclusions or analyze trends in public opinion, particularly in an area as complex as educational financial aid. There is also a cultural aspect to take in to account here: the U.S. is a country where sarcasm is a major cultural component. Analyzing data in the same methodological structure might be more effective in cultures that do not have the same sarcastic language culture.

**Recommendations**:

Future research in this area could benefit from incorporating a broader range of data sources and sentiment analysis tools, trained specifically on understanding sarcasm in social media. Expanding the dataset to include comments from various social media platforms could provide a more comprehensive view.

Despite its weaknesses, this sentiment analysis project has shed critical light on the public's perspective towards FAFSA, revealing a significant tilt towards negative sentiment when analyzing the raw data.

```python
import pandas as pd
import string
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
punctuation_table = str.maketrans('', '', string.punctuation)

def clean_and_lemmatize(text):
    text = text.lower().translate(punctuation_table)
    words = word_tokenize(text)
    lemmatized = [lemmatizer.lemmatize(word) for word in words if word not in stop_words and word.isalpha()]
    return ' '.join(lemmatized)

file_path = '/content/drive/MyDrive/Desktop/UOgIb7StyPU/youtube_comments.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    comments = file.read().splitlines()

cleaned_comments = [clean_and_lemmatize(comment) for comment in comments]

cleaned_comments_df = pd.DataFrame(cleaned_comments, columns=['cleaned_comment'])

output_file_path = '/content/drive/MyDrive/Desktop/UOgIb7StyPU/cleaned_youtube_comments.csv'
cleaned_comments_df.to_csv(output_file_path, index=False)

print(f'Cleaned comments have been saved to {output_file_path}')
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return nltk.corpus.wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return nltk.corpus.wordnet.VERB
    elif treebank_tag.startswith('N'):
        return nltk.corpus.wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return nltk.corpus.wordnet.ADV
    else:
        return nltk.corpus.wordnet.NOUN

def lemmatize_text(text):
    words = nltk.word_tokenize(str(text).lower())
    pos_tagged = pos_tag(words)
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tagged
        if word not in stop_words and word.isalpha()
    ]
    return ' '.join(lemmatized_words)
```

```python
from google.colab import drive
drive.mount('/content/drive')

!pip install nltk
import pandas as pd
import nltk
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import wordnet

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
punctuation_table = str.maketrans('', '', string.punctuation)

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_and_lemmatize_with_pos(text):
    if pd.isna(text):
        return ""
    text = str(text).lower().translate(punctuation_table)
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags if word.isalpha() and word not in stop_words]
    return ' '.join(lemmatized_words)

input_file_path = '/content/drive/MyDrive/Desktop/mTHtn0FRMWw/mTHtn0FRMWw_cleaned_comments.csv'
comments_df = pd.read_csv(input_file_path)

comments_df['Lemmatized_Comment'] = comments_df['Cleaned_Comments'].apply(clean_and_lemmatize_with_pos)

output_file_path = '/content/drive/MyDrive/Desktop/mTHtn0FRMWw/mTHtn0FRMWw_further_lemmatized_comments.csv'
comments_df.to_csv(output_file_path, index=False)

print(f"Further lemmatized comments saved to {output_file_path}")
```

```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_and_lemmatize(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    stop_words = set(stopwords.words('english'))
    words = [word for word in word_tokenize(text) if word not in stop_words]
    tagged = pos_tag(words)
    lemmatizer = WordNetLemmatizer()
    lemmatized_output = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in tagged]
    return ' '.join(lemmatized_output)

file_path = '/content/drive/MyDrive/Desktop/Pn4OECMTh5w/comments.txt'

with open(file_path, 'r') as file:
    comments = file.readlines()

cleaned_comments = [clean_and_lemmatize(comment) for comment in comments]

df_cleaned = pd.DataFrame(cleaned_comments, columns=['cleaned_comment'])

output_file_path = '/content/drive/MyDrive/Desktop/Pn4OECMTh5w/cleaned_comments.csv'
df_cleaned.to_csv(output_file_path, index=False)

print(f"Cleaned comments saved to {output_file_path}")
```
